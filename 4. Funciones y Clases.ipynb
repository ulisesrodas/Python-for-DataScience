## <a id = '#1.1.'> 1.1. Importancia de las funciones en Python </a> 

Una función es como una **caja** porque recibe un input y arroja un output. Nosotros especificamos lo que se quiere crear. Características de las funciones:
* **Abstracción y Reutilización:** Replica el código que usarás una y otra vez.

## <a id='#1.4.'>1.4. Múltiples objetos de retorno</a> 
* El output de una función puede tener varios objetos. Por defecto, se almacenan en una **tupla**.
* La tupla se almacena en paréntesis.

```python
def calculator_square( x, y ):
    
    x2 = x * x
    y2 = y * y
    
    result = x2 * y2   
    
    # 3 objetos: resultado y los cuadrados de los objetos iniciales
    return result, x2, y2
```

```python
# Defino que x = 3, y = 4 más directamente
calculator_square(3, 4)
```

```python
# Ahora, x = 5, y = 6
result4 = calculator_square(5, 6)
result4
```

## <a id='#1.6.'>1.6. Ejercicio </a> 

```python
# Base de datos de GitHub Innovation
# Observo la cantidad de programadores por tópicos por país

url = "https://raw.githubusercontent.com/github/innovationgraph/refs/heads/main/data/topics.csv"
topics_data = pd.read_csv(url)
topics_data
```

```python
# Quiero saber en cuántos países se discuten tópicos exclusivos de Deep Learning:
# Defino una función

def count_unique_countries_for_topic(df, target_topic="machine-learning"):

    # Filtra las filas donde el tópico sea igual al que queremos buscar:
    filtered_df = df[df['topic'] == target_topic]
    
    # Obtengo cuántos iso2_code (países únicos) únicos hay:
    unique_countries = filtered_df['iso2_code'].nunique()
    
    return unique_countries
```

```python
count_unique_countries_for_topic(df = topics_data, target_topic = "machine-learning")
```

```python
# Ahora queremos los top 5 países por número de programadores (num_pushers)
def top_countries_by_topic(df, target_topic="machine-learning", top_n=5):

    # Filtro por tópico:
    filtered_df = df[df['topic'].str.lower() == target_topic.lower()]
    
    # Agrupamos por país y sumamos los números de publicadores por país:
    country_pushers = filtered_df.groupby('iso2_code')['num_pushers'].sum()
    
    # Orden descendiente:
    top_countries = country_pushers.sort_values(ascending=False).head(top_n)
    return top_countries
```

```python
# Podemos cambiar el tópico a voluntad
top_countries_by_topic(topics_data, "machine-learning")
```

##  <a id='#2.6.'>2.6. \_\_init\_\_() </a>

En las clases de Python, el método “__init__” está reservado. Se llama automáticamente cuando creas un objeto usando una clase y se utiliza para inicializar las variables de dicha clase. Es equivalente a un constructor. 

* Al igual que cualquier otro método, el método init comienza con la palabra clave “def”.
* **“self”** es el primer parámetro en este método, al igual que en cualquier otro método, aunque en el caso de init, **“self”** se refiere a un objeto recién creado, a diferencia de otros métodos donde se refiere al objeto o instancia actual asociado con ese método en particular.
* Se pueden agregar parámetros adicionales.

## <a id='#2.7.'>2.7. Self</a>
* **self** es el mecanismo interno de Python que permite que un objeto diga 'yo' o 'mi' para acceder a sus propios datos y acciones sin confundirse con el resto
* Tres reglas clave para recordarlo:
    * No es una palabra reservada del sistema: Es una convención estricta de la comunidad de Python. Podrías llamarlo de otra forma, pero nadie lo hace.
    * Siempre va primero: En cualquier método dentro de una clase (incluido el constructor __init__), self tiene que ser obligatoriamente el primer parámetro dentro del paréntesis.
    * Es invisible al usarlo: Cuando tú, como usuario, llamas a un método (ej. mi_encuesta.calcular_estrato()), no escribes self adentro de los paréntesis. Python lo envía "por debajo de la mesa" de forma automática.

**Conectando Clases, __int__ y self:**
* Imaginemos que estamos administrando el trabajo de campo y recibo cuestionarios físicos que hay que digitalizar:
    * La Clase es el diseño de la base de datos (qué columnas tendrá).
    * __init__ es el momento exacto en que abres una nueva fila en blanco en tu base de datos para ingresar una nueva encuesta.
    * self es el cursor parpadeando en esa fila específica, listo para recibir los datos y asegurarse de que no se escriban en la fila de arriba ni en la de abajo.

* La secuencia lógica es la siguiente:
    * Cuando le dices a Python que cree un nuevo objeto a partir de tu Clase, el lenguaje necesita construirlo y llenarlo con sus primeros datos.
    * Para construirlo, Python llama automáticamente al método constructor: __init__.
    * Pero __init__ es solo una función ejecutándose; necesita saber a quién le va a asignar los datos. Por eso, Python genera el objeto vacío y se lo entrega a __init__ bajo el nombre de self.

## <a id='#2.8.'>2.8. Definiciones integradoras</a> 

* Cuando defines una **Clase** en Python, no estás procesando datos reales todavía. Simplemente le estás entregando un "manual de instrucciones" a la computadora. Le avisas al sistema: "A partir de ahora, existirá una nueva estructura de datos con estas reglas", y Python guarda ese diseño en su memoria general a la espera de que lo uses.

* La verdadera acción comienza con el **__init__**. Este es un método inicializador que entra a trabajar en el milisegundo exacto en que decides crear un caso o registro nuevo basándote en el manual de tu Clase. Python lo ejecuta de forma automática para darle nacimiento a ese registro.

* En el momento en que el **__init__** arranca, Python separa un bloque de memoria RAM exclusivamente para ese nuevo caso. Aquí es donde entra el **self**. El **self** es, literalmente, la dirección técnica o el "link" a ese bloque de memoria. Es un ancla que le dice a la computadora: "los datos que voy a recibir ahora no los dejes sueltos en el sistema, guárdalos exactamente en este espacio privado".

* Gracias a que tienes esa dirección de memoria asegurada por el self, puedes crear y guardar los **Atributos**. Por ejemplo, cuando ingresas la edad de un informante, Python viaja por el self hasta ese bloque de memoria reservado y anota el número de forma permanente. Si no usaras el self, la computadora olvidaría el dato apenas termine de procesar la creación del registro.

* Finalmente, tienes los **Métodos**. Estas son funciones que viven anidadas dentro de la **Clase** y que comparten el mismo **self**. Al tener esa "llave" de acceso a la memoria, los **Métodos** pueden entrar al espacio privado del registro, leer los atributos que guardaste (como la edad) y ejecutar acciones lógicas sobre ellos, como cambiar el estado de validación o lanzar una alerta si el dato no es consistente.

En el ejemplo, **MyFirstClass** usa self para definir atributos y métodos:

```python
class MyFirstCar:

# ATRIBUTOS: Llamos a la función int que nos permitirá generarlos
# Argumentos: self va por estructura, seguido de los atributos que queremos (pueden ir listas también)
# Función para definir Atributos:

    def __init__(self, model, year, color):

        # El atributo llamado modelo_car va a tomar el valor de modelo cuando coloquemos algún valor de modelo (model). Lo mismo con year_manu
        self.model_car = model
        self.year_manu = year
        self.color_car = color

# Hasta aquí, la clase MyFirstCar tiene tres atributos: model_car, year_manu y color_car
# Ahora, MÉTODOS:
# Función para definir Métodos: self va por estructura igualmente

# 1. Calcular cuántos años de antiguedad tiene el carro: creamos una función que le llamamos year_diff

    def year_diff(self): # Nos habilita a llamar a todos los argumentos, pero solo queremos el año de manufactura
        diff = 2026 - self.year_manu
        print("El año de diferencia es", diff)

# 2. Generamos una descripción:

    def description(self):
        print("Este carro es de modelo", self.model_car, "y fue fabricado en", self.year_manu)
```

**¿QUÉ SON LAS CLASES?**

**¿Cómo identificar Atributos y Métodos?**
* Todo lo que está debajo de la función __init__ son Atributos: se llama sin paréntesis.
* Lo que está debajo de otras funciones, son Métodos: se llama con punto y paréntesis.


```python
# Creo el objeto con los atributos que le especifico:
# También es válido: mycar = MyFirstCar(model = "Lexus", year = 2020, color = "beige")
mycar = MyFirstCar("Lexus", 2025, "beige")
```

```python
# En caso no esté importado
# import pandas as pd

class LanguageData:

    # Atributos:

    def __init__(self, data_source):
        if isinstance(data_source, str):
            self.data = pd.read_csv(data_source)
        else:
            self.data = data_source
            
        self.number_of_years = self.data['year'].nunique()
        self.number_of_countries = self.data['iso2_code'].nunique()
        self.number_of_unique_quarters = self.data['quarter'].nunique()

    # Métodos:   
    def get_programmers(self, country_code, language, year, quarter):
        """Get number of programmers for specific country, language, year, quarter"""
        result = self.data[
            (self.data['iso2_code'] == country_code) &
            (self.data['language'] == language) &
            (self.data['year'] == year) &
            (self.data['quarter'] == quarter)
        ]
        return result['num_pushers'].iloc[0] if not result.empty else None
```

```python
# Ejemplo
url = "https://raw.githubusercontent.com/github/innovationgraph/refs/heads/main/data/languages.csv"
analyzer = LanguageData(url)
```

```python
# Atributo: Llamamos a los datos
analyzer.data
```

```python
# Atributo
analyzer.number_of_years
```

```python
# Atributo
analyzer.number_of_countries
```

```python
# Atributo
analyzer.number_of_unique_quarters
```

```python
# Método: Usamos paréntesis para invocarlo
# Número de programadores en HTML en Perú durante el 2020, quarter 3
analyzer.get_programmers("PE", "HTML", 2020, 3)
# Salida: 8854
```
