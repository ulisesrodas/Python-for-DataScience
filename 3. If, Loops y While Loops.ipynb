# <a id='1'> Condifición If </a>

## <a id='1.1'> 1. La estructura del código</a> 


La función **if** evalúa condiciones: 
- Igual: a == b
- No igual: a != b
- Menor que: a < b
- Menor o igual que: a <= b
- Mayor que: a > b
- Mayor o igual que: a >= b

Si se cumple la condición, sigue. Si no, se saltea.

```python
# Ejemplo: El resultado es un valor booleano.
A = 20
B = 30
resultado = A > B
resultado
```

```python
# Cuando la condicionante sea verdad, se ejecutarán las instrucciones dentro del bloque if.
# Si no es verdad, no se ejecuta nada.
Economía = 2014
Sociología = 2019
```

```python
# No se ejecuta nada porque no se cumple la condición.
# Todo lo que esté dentro del bloque if (tiene tabulación) no se ejecutará.
if Economía > Sociología:
    print("Se ingresó primero a Economía")
    print("Aunque no se terminó la carrera")
```

```python
# Puntaje de las universidades peruanas en rankings
PUCP = 1022
UNMSM = 813
```

```python
# Usamos 'else' para explicitar qué va a ocurrir si la condición no se cumple.
if UNMSM > PUCP:
    print("UNMSM es mejor que PUCP")
else:
    print("PUCP es la mejor universidad del Perú")
```

* La función **if** mide la veracidad de una expresión lógica. El resultado es ser un **<font color='red'>boolean</font>**. En otras palarbas, el output es **<font color='red'>True</font>** or **<font color='red'>False</font>**.

# <a id='2'> Loops </a> 

* Su premisa principa es acelerar un proceso que consideramos repetitivo. 
* Un bucle for se utiliza para iterar sobre una secuencia. Ejecuta un bloque de código un número definido de veces.



## <a id='2.1'> 1. In NumPy </a> 

```python
import numpy as np
```

```python
# Elevemos al cuadrado manualmente cada elemento de la lista:
sequence_ages1 = [ 26, 35, 42, 32, 41]
```

```python
al_1 = 26**2
al_2 = 35**2
al_3 = 42**2
al_4 = 30**2
al_5 = 41**2

print(al_1)
print(al_2)
print(al_3)
print(al_4)
print(al_5)
```

```python
# Restar 10 a cada elemento:
# Podemos usar cualquier letra o letras como iterador: i, age, etc.
# Es mejor usar un iterador relacionado a lo que se hace
for age in sequence_ages2:
    print(age - 10)
```

```python
# Multiplicamos *2
for age in sequence_ages2:
    print(age*2)
```

## <a id='2.2'>2. Range </a>
* Tener en cuenta que Python no toma el último valor del range(n)

* Para crear secuencias, Loop y Range es lo mismo

```python
# El 'for' es el loop (la repetición)
# El 'range(5)' genera la secuencia secreta: 0, 1, 2, 3, 4
for i in range(5):
    print(i)
```

* Diferencias del for con el range: 
  * El loop (como el for) es el "motor" o la acción de repetir un bloque de código.
  * El range() es una función nativa de Python (técnicamente, un tipo de secuencia) que genera una lista de números. Actúa como el "combustible" o la "pista" que le dice al loop cuántas vueltas tiene que dar.

* El range por sí solo no repite ninguna acción, solo crea una "colección" de números lista para ser usada.
* El bucle for no necesita un range para funcionar. Su trabajo es simplemente recorrer cualquier cosa que tenga varios elementos (como una lista de textos).
* Normalmente los usamos juntos cuando queremos repetir un bloque de código una cantidad exacta de veces, pero sin tener que escribir una lista a mano.
* En conclusión: range() es el objeto que contiene los números, y el bucle for es la instrucción que ejecuta una acción repetitiva.

```python
# Solo Range: 
# No repite nada, solo dice lo que es: 
mis_numeros = range(5)
print(mis_numeros) 
```

```python
# Solo for:
# El bucle 'for' es la acción de recorrer la lista
nombres = ["Ulises", "Flor de María", "Genaro", "María"]
for persona in nombres:
    print("Hola", persona)
```

```python
# Juntamos Range y for:
# El 'for' hace el trabajo de repetir.
# El 'range(5)' le da los 5 "boletos" (0, 1, 2, 3, 4) para que sepa cuántas veces hacerlo.
for numero in range(5):
    print("Ejecutando vuelta número:", numero)
```

# <a id='4'> While Loop </a>
* While loop se utiliza para repetir código mientras la condición sea verdadera. Cuando la expresión lógica es falsa, el código terminará su ejecución.
* Diferencias con Loop:
  * Mientras que el Loop acaba cuando se llega al último elemento
  * El WhileLoop se acaba cuando el stament no se cumpla más, por eso el corazón de su estructura es un **break code**: algo que pueda romper el código.


## <a id='4.4.1.'> 1. Estructura </a>

while <font color='green'>condition</font>: <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;code 1<br><br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<font color='red'>break code</font><br><br>

* El operador break se utiliza para interrumpir y salir inmediatamente de un bucle (como un while o un for), sin importar si la condición principal del bucle sigue siendo verdadera.
* Es indispensable su presencia, de lo contrario seguirá operando hasta el infinito.

```python
i = 1

while i < 6 : 
    print ("Soy un estudiante de kinder porque mi edad es", i)
    
    # break code
    # Mientras la condición sea verdadera, se ejecutará el bloque de código dentro del while
    # Específicamente, la "i" debe iterar hasta 5, dado que i+1 es 6. Cuando llegue ese momento, se detiene (rompe) el código.
    i = i + 1   

```

* El While Loop es más usado cuando hay web scraping y no se conoce el fin de las páginas a revisar.
* Si sé que hay 5 páginas, uso un Loop solamente.

```python
i = 1

while i < 10 :
    
    print(i+10)
    
    # Break code 
    i = i + 5
```
